**Food Delivery App Data Analysis**
<br>Zomato is one of the most used food ordering platforms. 
<br>This project aims on cleaning the dataset, analyze the given dataset, and mining informational quality insights.



```python
import pandas as pd
import numpy as np
```

```python
hotels = pd.read_csv('zomato.csv')
```

```python
hotels.head(10)
```

```python
hotels.info()
```

```python
# Remove unwanted columns
hotels = hotels.drop(['address', 'phone'], axis=1)
```

```python
hotels.head()
```

```python
# Rename columns
hotels = hotels.rename(columns={
                        'rate': 'rating',
                        'approx_cost(for two people)': 'approx_cost',
                        'listed_in(type)': 'type'})
hotels.head()
```

```python
hotels.dtypes
```

```python
 #deleting null values of name column
hotels = hotels.dropna(subset=['name'])
```

```python
 #handling null values of book_table
hotels['book_table'] = hotels['book_table'].fillna(value = 'NA')
```

```python
 #handling null values of rating
hotels['rating'] = hotels['rating'].fillna(value = 0)
```

```python
#handling null values of votes
hotels['votes'] = hotels['votes'].fillna(value = 0)
```

```python
#handling null values of location
hotels['location'] = hotels['location'].fillna(value = 'NA')
```

```python
#handling null values of rest_type
hotels['rest_type'] = hotels['rest_type'].fillna(value = 'NA')
```

```python
#handling null values of dishliked
hotels['dish_liked'] = hotels['dish_liked'].fillna(value = 'NA')
```

```python
#handling null values of cuisines
hotels['cuisines'] = hotels['cuisines'].fillna(value = 'NA')
```

```python
#handling null values of approxcost
hotels['approx_cost'] = hotels['approx_cost'].fillna(value = 0)
```

```python
#handling null values of type
hotels['type'] = hotels['type'].fillna(value = 'NA')
```

```python
#handling null values of online_order
hotels['online_order'] = hotels['online_order'].fillna(value = 'NA')
```

```python
hotels.isnull().sum()
```

```python
#droping the duplicates value keeping the first
hotels = hotels.drop_duplicates()
hotels.head()
```

```python
#removing irrelevant text from all the columns
for col in hotels.columns:
    hotels = hotels[hotels[col].str.match(r'^.*?(RATED|Rated).*$')==False]
```

```python
hotels.head(10)
```

```python
hotels.shape
```

```python
# Check for unique values in each column and handle the irrelevant values
hotels = hotels[hotels['online_order'].str.match(r'^(No|Yes)$')==True]
hotels['rating'] = hotels['rating'].str.replace('/5', '')
def replace_rating(rating):
    if rating == 'NEW' or rating == '-':
        return 0
    else:
        return rating
hotels['rating'] = hotels['rating'].apply(replace_rating)
```

```python
hotels['rating'].head(10)
```

```python
# Remove the unknown character from the column 'name'
hotels['name'] = hotels['name'].str.replace('[Ãx][^A-Za-z]+','',regex=True)
```

```python
# Export dataframe to "zomatocleaned.csv"
hotels.to_csv('zomatocleaned.csv')
```
